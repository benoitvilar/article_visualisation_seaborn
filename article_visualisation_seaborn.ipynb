# 2 fonctionnalités pour améliorer les visualisations sur SEABORN

Pour pouvoir analyser des données, l’étape de visualisation est essentielle afin de trouver les bons axes d’étude. Parmi les différentes librairies présentes sous Python, de nombreux types de visualisation existent mais encore faut-il bien connaître les différents possibilités. Certaines fonctionnalités de SEABORN sont moins utilisées mais sont très utiles.

# PAIRPLOT

PAIRPLOT permet de visualiser les corrélations entre les différents éléments d’un dataset par l’intermédiaire d’une seule instruction au lieu de plusieurs avec d’autres fonctions.
En prenant en exemple un dataset représentant les dépenses et les profits de startup dans 3 villes américaines, voici les différentes représentations graphiques que nous obtenons.

```python
#import des librairies
import seaborn as sns
import pandas as pd
```

```python
#import du dataset
startup = pd.read_csv('50_Startups.csv')
```

```python
#visualisation des premières lignes du dataset
startup.head()
```

```python
#visualisation des corrélations entre les variables via la fonction pairplot de Seaborn
sns.pairplot(startup, hue='State', height=2.5)
```

# JOINPLOT

JOINPLOT permet de visualiser différentes informations concernant l’analyse de deux variables. Le paramètre “kind” offre la possibilité d’observer des densités (“kde”), des hexagones (“hex”) ou des régressions (“reg”).

```python
#import du dataset
house = pd.read_csv('house_pricing.csv')
```

```python
#visualisation de la distribution des prix selon la superficie via la fonction pairplot de Seaborn
g=sns.jointplot(x='LotArea',y='SalePrice',data=house)
```

La visualisation doit être améliorée en enlevant les valeurs extrêmes (LotArea>20000 et SalePrice>300000)

```python
#nouveau dataset
house1=house.loc[(house['SalePrice']<300000)&(house['LotArea']<20000)]
```

```python
#visualisation de la distribution des prix selon la superficie via la fonction pairplot de Seaborn
g=sns.jointplot(x='LotArea',y='SalePrice',data=house1)
```

```python
#ajout de la valeur 'reg' pour le paramètre kind 
g=sns.jointplot(x='LotArea',y='SalePrice',data=house1,kind='reg')
```

```python
#ajout de la valeur 'hex' pour le paramètre kind 
g=sns.jointplot(x='LotArea',y='SalePrice',data=house1,kind='hex')
```

```python
#ajout de la valeur 'kde' pour le paramètre kind 
g=sns.jointplot(x='LotArea',y='SalePrice',data=house1,
                  kind="kde", space=0, color="g")
```
